```python
df = sns.load_dataset("titanic")
sns.violinplot(x=df["age"])
```

```python
resultado=df["result"]
edad=df ["age"]
fig,ax=plt.subplots()
ax.scatter(edad,resultado, marker='*',c='salmon')
plt.title('Gráfica de edad versus resultados')         
plt.xlabel('edad')                                           
plt.ylabel('resultado');
```

```python
resultado=df["result"]
edad=df ["age"]
fig,ax=plt.subplots()
ax.plot(resultado,edad, linestyle= '--')
```

```python
resultado=df["result"]
edad=df ["age"]
a9=df ["A9_Score"]
fig,ax=plt.subplots()
ax.stackplot(edad,resultado,a9)
plt.title('Gráfica de edad versus resultados y A9')         
plt.xlabel('edad')                                           
plt.ylabel('resultado');
```
